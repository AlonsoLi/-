```python
#让两部分显示不同的颜色
x= np.random.rand(10)
x=x-x.mean()
x.sort()

#创建颜色表
colors=[]
for i in x:
    if i > 0:
        colors.append('red')
    else:
        colors.append('green')
        
plt.figure(figsize=(8,4))
plt.hlines(y=range(10),xmin=0,xmax=x
          ,linewidth=5
          ,color=colors
          ,alpha=0.5);#调整色饱和度
```

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
df.head(10)
```

```python
df.shape

#判断特征cars与carname的区别
sum(df.cars==df.carname)
```

```python
#基础图形

plt.figure(figsize=(12,8))

plt.hlines(y=df.cars,xmin=0,xmax=df.mpg
          ,color='b'
          ,alpha=0.5
          ,linewidth=5);
```

```python
#对目标数据进行标准化处理

x=df.loc[:,'mpg']
df['mpg_z']=(x-x.mean())/x.std()
df['mpg_z'].values
```

```python
#生成颜色标签
df['colors']=['green' if i>0 else 'red' for i in df['mpg_z']]
```

```python
#直接替代
color=[]
for i in df['mpg_z']:
    if i > 0:
        color.append('green')
    else:
        color.append('red')
print(color)
```

```python
df.head()
```

```python
#原数据是无序排列的，我们需要对数据进行排序
#这里需要注意的是，我们不止要对x进行排序，还需要确保y索引跟着一起变化
#所以我们应该对整个数据集进行排序
df.sort_values('mpg_z',inplace=True)
df.head()
```

```python
#排序之后索引也会跟着一起打乱
#需要重置索引
df.reset_index(drop=True,inplace=True) #drop参数是不要增加新的索引
df.head()
```

```python
#画图
plt.figure(figsize=(14,10),dpi=65)

plt.hlines(y=df.cars,xmin=0,xmax=df.mpg_z
          ,color=df.colors
          ,alpha=0.4
          ,linewidth=5)

#添加装饰
plt.ylabel('$model$',fontsize=15)   #设置y轴标签，fontsize调节文字大小 美刀变成斜体
plt.xlabel('$mileage$',fontsize=15) #设置x轴标签，fontsize调节文字大小
plt.yticks(fontsize=12)         #y轴刻度标签
plt.title('Diverging Bars of Car Mileage', fontdict={'size':20}) #设置图形标题 fontdict 字典型的参数
plt.grid(linestyle='--', alpha=0.5);  #配置网格线

```

```python
#导入Kmeans库
from sklearn.cluster import KMeans
```

```python
#导入原数据集
df1 = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
```

```python
#提取出目标数据，由于sklearn只接受二维数据，所以需要对目标数据进行变形
data_x = df1.mpg.values.reshape(-1,1)
print(data_x)
```

```python
#执行聚类，分成两类
cluster = KMeans(n_clusters=2, random_state=0).fit(data_x)
```

```python
#查看聚好的类别，每个样本所对应的类
cluster.labels_
```

```python
df1['label'] = cluster.labels_
df1.head()
```

```python
#对数据集进行排序
df1.sort_values('mpg', inplace=True)
df1.index = range(df1.shape[0])#另一种更改index的方式
df1.head()
```

```python
(df['label'] == df1['label']).sum()
```

```python
#计算准确率
(df['label'] == df1['label']).mean()
```
